```python
# Binary Columns
application_df["Gender"].replace(['F','M'], [1,0], inplace=True)
application_df["Own Car"].replace(['Y','N'], [1,0], inplace=True)
application_df["Own Realty"].replace(['Y','N'], [1,0], inplace=True)
application_df["Approval Status"].replace(['Approved','Rejected'], [1,0], inplace=True)
```
